```python
# Replace with your file path (use raw string or double backslashes for Windows)
file_path = r'/Users/constantin/Documents/Python_files/Machine_Learning/EY Challenge/Storm Damage - 2024/results/training/dataset.csv'  # Windows example
# file_path = '/home/user/path/to/your/file.csv'  # Linux/Mac example

# Read the CSV into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to verify
print(df.head())
```

```python
df['F1_val'] = 2 * (df['P_val'] * df['R_val']) / (df['P_val'] + df['R_val'])
df['F1_test'] = 2 * (df['P_test'] * df['R_test']) / (df['P_test'] + df['R_test'])
```

```python
import matplotlib.pyplot as plt

# Example: Plot 'temperature' and 'humidity' vs. 'timestamp'
plt.figure(figsize=(10, 5))  # Set figure size

# Plot the columns
# plt.plot(df['size_ds'], df['mAP_50_val'], label='mAP_50_validation') # , color='blue'
# plt.plot(df['size_ds'], df['mAP_50_test'], label='mAP_50_test')
plt.plot(df['size_ds'], df['P_val'], label='Precision validation')
plt.plot(df['size_ds'], df['P_test'], label='Precision test')
plt.plot(df['size_ds'], df['R_val'], label='Recall validation')
plt.plot(df['size_ds'], df['R_test'], label='Recall test')
# plt.plot(df['size_ds'], df['F1_val'], label='F1 validation')
# plt.plot(df['size_ds'], df['F1_test'], label='F1 test')


# Customize the plot
plt.title('Temperature and Humidity Over Time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()  # Show labels
plt.grid(True)  # Add grid lines

# Rotate x-axis labels if they overlap
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
